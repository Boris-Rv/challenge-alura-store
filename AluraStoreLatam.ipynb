### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
suma_total = tienda['Precio'].sum()
suma_total2 = tienda2['Precio'].sum()
suma_total3 = tienda3['Precio'].sum()
suma_total4 = tienda4['Precio'].sum()
```

```python
print(f'Ingreso total de la tienda 1 es: {suma_total}')
print(f'Ingreso total de la tienda 2 es: {suma_total2}')
print(f'Ingreso total de la tienda 3 es: {suma_total3}')
print(f'Ingreso total de la tienda 4 es: {suma_total4}')
```

```python
print(tienda.shape)
print(tienda2.shape)
print(tienda3.shape)
print(tienda4.shape)
```

# 2. Ventas por categoría

```python
vpt1 = tienda.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending=False)
vpt2 = tienda2.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending=False)
vpt3 = tienda3.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending=False)
vpt4 = tienda4.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending=False)
```

```python
print("VENTAS POR CATEGORÍA")
print("=" * 70)

print('tienda 1')
print(f'{vpt1}')
print('\ntienda 2')
print(f'{vpt2}')
print('\ntienda 3')
print(f'{vpt3}')
print('\ntienda 4')
print(f'{vpt4}')
```

# 3. Calificación promedio de la tienda


```python
def calificacion_promedio(tienda,numero):
    promedio = tienda['Calificación'].mean()
    print(f'Calificacion promedio por tienda {numero}: {promedio:.2f} ')


calificacion_promedio(tienda, 1)
calificacion_promedio(tienda2, 2)
calificacion_promedio(tienda3, 3)
calificacion_promedio(tienda4, 4)
```

# 4. Productos más y menos vendidos

```python
# Para Tienda 1
productos_t1 = tienda.groupby('Producto')['Precio'].sum()
print('Tienda 1:')
print(f'  Más vendido: {productos_t1.idxmax()} (${productos_t1.max():,.0f})')
print(f'  Menos vendido: {productos_t1.idxmin()} (${productos_t1.min():,.0f})')

# Para Tienda 2
productos_t2 = tienda2.groupby('Producto')['Precio'].sum()
print('\nTienda 2:')
print(f'  Más vendido: {productos_t2.idxmax()} (${productos_t2.max():,.0f})')
print(f'  Menos vendido: {productos_t2.idxmin()} (${productos_t2.min():,.0f})')

# Para Tienda 3
productos_t3 = tienda3.groupby('Producto')['Precio'].sum()
print('\nTienda 3:')
print(f'  Más vendido: {productos_t3.idxmax()} (${productos_t3.max():,.0f})')
print(f'  Menos vendido: {productos_t3.idxmin()} (${productos_t3.min():,.0f})')

# Para Tienda 4
productos_t4 = tienda4.groupby('Producto')['Precio'].sum()
print('\nTienda 4:')
print(f'  Más vendido: {productos_t4.idxmax()} (${productos_t4.max():,.0f})')
print(f'  Menos vendido: {productos_t4.idxmin()} (${productos_t4.min():,.0f})')

```

#Generando Gráfico

```python
from matplotlib import pyplot as plt
```

```python


tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
Ingresos = [suma_total, suma_total2, suma_total3, suma_total4]

plt.figure(figsize=(8,6))
plt.bar(tiendas, Ingresos, color=['#3498db', '#3498db', '#3498db', '#e74c3c'],
        edgecolor='black', linewidth=1.2)
plt.title('Ingreso total por tienda', fontsize=16)
plt.xlabel('Tiendas')
plt.ylabel('Ingresos ($)')
for i, valor in enumerate(Ingresos):
    plt.text(i,
             valor + (max(Ingresos)*0.01),
             f'${valor:,.2f}',
             ha='center',
             va='bottom',
             fontsize=10,
             fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
tienda['N_tienda'] = 'Tienda 1'
tienda2['N_tienda'] = 'Tienda 2'
tienda3['N_tienda'] = 'Tienda 3'
tienda4['N_tienda'] = 'Tienda 4'
df_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Crear tabla pivote directamente desde df_tiendas
df_categorias = df_tiendas.pivot_table(
    index='Categoría del Producto',
    columns='N_tienda',
    values='Precio',
    aggfunc='sum',
    fill_value=0
)

# Ordenar por la suma total de todas las tiendas
df_categorias = df_categorias.loc[df_categorias.sum(axis=1).sort_values(ascending=False).index]

# Colores
colors_by_shop = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Graficar
df_categorias.plot(kind='bar', color=colors_by_shop, figsize=(12, 7))
plt.title("Ventas por categoría", fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Ventas ($)', fontsize=12)
plt.xlabel('Categorías de Producto', fontsize=12)
plt.legend(title='Tienda', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

```python

Tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
Promedio = [tienda['Calificación'].mean(), tienda2['Calificación'].mean(),
            tienda3['Calificación'].mean(), tienda4['Calificación'].mean()]

plt.figure(figsize=(10,6))
plt.plot(["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"], Promedio, marker='o')
plt.title("Calificación Promedio por Tienda")
plt.ylabel("Promedio de Calificación")
plt.xlabel("Tienda")
plt.grid(axis='y')
plt.show()
```

```python
# Primero, vamos a recolectar todos los datos
datos_tiendas = []

# Tienda 1
productos_t1 = tienda.groupby('Producto')['Precio'].sum()
datos_tiendas.append({
    'tienda': 'Tienda 1',
    'mas_vendido': productos_t1.idxmax(),
    'valor_mas': productos_t1.max(),
    'menos_vendido': productos_t1.idxmin(),
    'valor_menos': productos_t1.min()
})

# Tienda 2
productos_t2 = tienda2.groupby('Producto')['Precio'].sum()
datos_tiendas.append({
    'tienda': 'Tienda 2',
    'mas_vendido': productos_t2.idxmax(),
    'valor_mas': productos_t2.max(),
    'menos_vendido': productos_t2.idxmin(),
    'valor_menos': productos_t2.min()
})

# Tienda 3
productos_t3 = tienda3.groupby('Producto')['Precio'].sum()
datos_tiendas.append({
    'tienda': 'Tienda 3',
    'mas_vendido': productos_t3.idxmax(),
    'valor_mas': productos_t3.max(),
    'menos_vendido': productos_t3.idxmin(),
    'valor_menos': productos_t3.min()
})

# Tienda 4
productos_t4 = tienda4.groupby('Producto')['Precio'].sum()
datos_tiendas.append({
    'tienda': 'Tienda 4',
    'mas_vendido': productos_t4.idxmax(),
    'valor_mas': productos_t4.max(),
    'menos_vendido': productos_t4.idxmin(),
    'valor_menos': productos_t4.min()
})

# Crear gráfico
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Productos Más y Menos Vendidos por Tienda', fontsize=16, fontweight='bold')

for idx, datos in enumerate(datos_tiendas):
    ax = axes[idx//2, idx%2]

    # Datos para este subgráfico
    productos = [datos['mas_vendido'], datos['menos_vendido']]
    valores = [datos['valor_mas'], datos['valor_menos']]
    colores = ['#2ecc71', '#e74c3c']  # Verde para más vendido, rojo para menos

    # Crear barras
    bars = ax.bar(productos, valores, color=colores, edgecolor='black')

    # Personalizar
    ax.set_title(f"{datos['tienda']}", fontsize=14, fontweight='bold')
    ax.set_ylabel('Ingresos ($)', fontsize=11)
    ax.grid(axis='y', linestyle='--', alpha=0.6)

    # Añadir valores en las barras
    for bar, valor in zip(bars, valores):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, height + (max(valores)*0.01),
                f'${valor:,.0f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
Tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
Promedio = [tienda['Costo de envío'].mean().astype(float), tienda2['Costo de envío'].mean().astype(float),
            tienda3['Costo de envío'].mean().astype(float), tienda4['Costo de envío'].mean().astype(float)]


plt.figure(figsize=(10,6))
plt.pie(Promedio,
        labels=Tiendas,
        autopct='%1.1f%%',
        startangle=90,
        colors=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])

plt.title('Envio Promedio por tienda', fontsize=16, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
